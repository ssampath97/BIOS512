```r
# uncomment to load the libraries

library("tidyverse")
library("patchwork")
options(repr.plot.res = 300)
```

```r
p1 = ggplot(mpg, aes(x=displ, y=hwy))
p1 = p1+geom_point()

p2 = ggplot(iris, aes(x=Sepal.Length, y=Sepal.Width, color=Species))
p2 = p2+geom_point()

p1+p2
```
