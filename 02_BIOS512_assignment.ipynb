Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
options(repr.plot.width=8, repr.plot.height=6)
p = ggplot(penguins, aes(x=body_mass_g, group = year))
p = p + facet_wrap(~year, ncol=1, scales = 'free_y')
p = p + geom_histogram(bins=35, color = "blue", alpha = 0.2) 
p = p + theme_linedraw()
p = p + scale_x_continuous(labels = comma_format())
p
```

Is there a difference between the body mass of male and female penguis? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to see compare within each species-group.**

```r
options(repr.plot.width=4, repr.plot.height=6)
p = ggplot(penguins, aes(x = body_mass_g, y = sex))
p = p + geom_point(size = 1, position = position_jitter(w=0, h=0.2), alpha=0.5)
p = p + facet_wrap(~species, ncol=1, scales = "free_y")
p
```
